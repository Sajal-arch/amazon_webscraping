```python
import requests
from bs4 import BeautifulSoup
import csv

# Making a GET request
r = requests.get('https://www.amazon.in/s?k=samsung+mobile&crid=3124USQN74ES&sprefix=Samsung+%2Caps%2C271&ref=nb_sb_ss_ts-doa-p_2_8')

# Parsing the HTML
soup = BeautifulSoup(r.text, 'html.parser')

mobile_listings = soup.find_all('div', {'class': 'puisg-row'})
print(mobile_listings)

filename = 'titles.csv'
with open(filename, 'a', newline='') as f:
    w = csv.DictWriter(f,['Price','Amount','Description'])
    w.writeheader()

for listing in mobile_listings:
    d ={}
    titles_list = []
    # Find the elements containing mobile details
    price_element = listing.find('span', {'class': 'a-price-whole'})
    amount_element = listing.find('span',{'class': 'a-price a-text-price'})
    description_element = listing.find('span', {'class': 'a-size-medium a-color-base a-text-normal'})

    # Extract text from the elements (assuming text is directly visible)
    price = price_element.text.strip() if price_element else 'Not found'
    amount = amount_element.text[1:7].strip() if amount_element else 'Not found'
    description = description_element.text.strip() if description_element else 'Not found'

    if description != 'Not found':
        d['Price'] = price
        d['Amount'] = amount
        d['Description'] = description
        titles_list.append(d)
        with open(filename, 'a', newline='') as f:
            w = csv.DictWriter(f,['Price','Amount','Description'])
            w.writerows(titles_list)

        # Print or process the scraped details
        print(f"Price: {price}, Amount: {amount}, Description: {description}")
```
